# 回帰タスクの評価指標
## 平均二乗誤差（MSE）
各サンプルの誤差の二乗の平均  
大きな誤差に対してより厳しくペナルティを課すため、外れ値の影響が大きい  
MSEが小さいほど、予測が実際の値に近い
## 平均絶対誤差（MAE）
各サンプルの誤差の絶対値の平均  
外れ値の影響がMSEほど強くなく、誤差の大きさを直感的に捉えやすい  
MAEが小さいほど、全体の予測誤差が小さい  
## RMSE（Root Mean Squared Error）
MSEの平方根  
元のデータの単位で表されるため、解釈しやすい  
RMSEも小さいほど、良い予測性能
## 決定係数（R²）
モデルがどれだけ実際のデータのばらつきを説明できているかを示す  
1に近いほどモデルの予測性能が高い。  
0の場合、平均値で予測した場合と同じ性能  
負の値になると、モデルの予測が非常に悪いことを示す　　
R²が高ければ、モデルがデータの変動をうまく捉えていると判断できる

```python
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt

# データ生成（人工的な回帰データ）
np.random.seed(42)
# 実際の値：0から100までの等間隔の値
y_true = np.linspace(0, 100, 100)
# 予測値：実際の値にランダムなノイズを加えたもの
y_pred = y_true + np.random.normal(scale=10, size=100)

# 評価指標の計算
mse = mean_squared_error(y_true, y_pred)           # 平均二乗誤差
mae = mean_absolute_error(y_true, y_pred)            # 平均絶対誤差
rmse = np.sqrt(mse)                                  # RMSE
r2 = r2_score(y_true, y_pred)                        # 決定係数 R²

# 結果の表示
print(f"MSE: {mse:.4f}")
print(f"MAE: {mae:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"R²: {r2:.4f}")

# 結果のプロット（実際の値と予測値の比較）
plt.plot(y_true, label='actual data')
plt.plot(y_pred, label='predict', linestyle='--')
plt.xlabel("sample number")
plt.ylabel("value")
plt.legend()
plt.show()
```
