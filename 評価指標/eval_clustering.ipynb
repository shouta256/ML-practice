# クラスタリングの評価指標
## シルエットスコア (Silhouette Score)
各サンプルについて同じクラスタ内の平均距離と最も近い他のクラスタまでの平均距離との差をどれだけクラスタ間の分離ができているかを評価する指標  
範囲は-1から1  
1に近いほど、クラスタ内部が凝集しており、クラスタ間が十分に分離されていることを示す
## ダビーズ・ボルディン指数 (Davies-Bouldin Score)
各クラスタの内部の凝集度と、クラスタ間の分離度を比較して評価する指標  
値が低いほど、クラスタ内部のデータが密集していて、クラスタ間の距離が広いことを意味する


```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
import matplotlib.pyplot as plt

# サンプルデータの生成：300サンプル、3つのクラスタ
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)

# KMeansクラスタリングの実行
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

# 評価指標の計算
silhouette = silhouette_score(X, labels)
davies_bouldin = davies_bouldin_score(X, labels)

print("Silhouette Score:", silhouette)
print("Davies-Bouldin Score:", davies_bouldin)

# クラスタリング結果の可視化
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.title("KMeans Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
