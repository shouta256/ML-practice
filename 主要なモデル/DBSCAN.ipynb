# DBSCAN（Density-Based Spatial Clustering of Applications with Noise）
密度に基づいてクラスタを検出する手法。  
密度が高い領域（高密度な点が集まった領域）をクラスタとして認識し、密度が低い領域はノイズとして除外する。 
## アルゴリズム
1.ε（イプシロン）とminPtsを設定する：
　　・ε: 半径ε内に含まれるデータポイントの数（密度の閾値）。
　　・minPts: クラスタを形成するために必要な最小のデータポイント数。
2.任意のデータポイントを選び、その点からε範囲内のすべての点を収集。
3.ε範囲内にminPts以上の点があれば、それをクラスタとして認識。
4.クラスタの内部で再帰的に隣接する点を収集し、クラスタを広げる。
5.密度が低い点は「ノイズ」として分類される。
## 特徴
クラスタ数を事前に指定しない（自動的にクラスタ数を決定）。  
密度の変化に強く、ノイズの多いデータにも対応。  
データが高次元になると、計算が複雑になることがある。  
密度の異なるクラスタがある場合やノイズが多い場合に強力。  

```python
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt

# サンプルデータ
X = np.random.rand(100, 2)

# DBSCANクラスタリング
dbscan = DBSCAN(eps=0.1, min_samples=5)
dbscan.fit(X)

# 結果の表示
plt.scatter(X[:, 0], X[:, 1], c=dbscan.labels_)
plt.show()
```
