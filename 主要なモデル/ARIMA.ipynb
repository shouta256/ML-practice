# ARIMA（AutoRegressive Integrated Moving Average）
時系列データの予測に広く使われる統計的なモデル  
過去のデータを基に未来の値を予測するために自己回帰（AR）、移動平均（MA）、積分（I）という3つの成分を組み合わせている
## 特徴
自己回帰（AR）: 過去の観測値に基づいて予測する成分。  
移動平均（MA）: 過去の予測誤差を使って予測する成分。  
積分（I）: 非定常な時系列データを定常化するための差分を取る成分。
## 用途
過去のデータが比較的線形で安定している場合に効果的
## モデルの構成
AR(p): 自己回帰の次数（過去p期のデータを使う）  
I(d): 差分化の次数（d回差分を取る）  
MA(q): 移動平均の次数（過去q期の誤差を使う）  


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# サンプルデータ（単純な時系列データ）
np.random.seed(42)
data = np.random.randn(100).cumsum()  # ランダムウォーク

# ARIMAモデルのフィット
model = ARIMA(data, order=(1, 1, 1))  # AR(1), I(1), MA(1)
model_fit = model.fit()

# 予測
forecast = model_fit.forecast(steps=10)

# 結果のプロット
plt.plot(data, label='actual data')
plt.plot(np.arange(len(data), len(data) + 10), forecast, label='predict', color='red')
plt.legend()
plt.show()
```
