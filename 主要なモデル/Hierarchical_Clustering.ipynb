# 階層的クラスタリング（Hierarchical Clustering）
クラスタをツリー構造（デンドログラム）で表現する手法  
階層的クラスタリングには凝集型（Agglomerative）と分割型（Divisive）の2つのアプローチがある  
一般的に使用されるのは凝集型
### 凝集型
各データを最初は個別のクラスタとして扱い、クラスタ同士を次々に統合していく
### 分割型
一つの大きなクラスタから始めて、最も異なるクラスタを次々に分割していく
## アルゴリズム
1.各データポイントを1つのクラスタとして扱う  
2.最も似ているクラスタをペアにして統合する（距離が最小になるように）。  
3.統合したクラスタ同士を再びペアにし、統合を繰り返す。  
4.最後に1つのクラスタに統合されるまで続ける。  
## 特徴
ツリー構造（デンドログラム）で結果が得られるため、クラスタ数を事前に指定しなくても良い  
複雑なデータに対して効果的  
凝集型アルゴリズムは計算量が O(n^3) で、データ数が多い場合は計算が遅くなる

```python
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import matplotlib.pyplot as plt

# サンプルデータ
X = np.random.rand(100, 2)

# 階層的クラスタリング
hierarchical = AgglomerativeClustering(n_clusters=3)
labels = hierarchical.fit_predict(X)

# 結果の表示
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.show()
```
