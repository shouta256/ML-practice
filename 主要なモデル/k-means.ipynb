# k-meansクラスタリング
あらかじめ指定した k 個のクラスタにデータを分ける手法
## アルゴリズム
クラスタ数 k を指定。  
データからランダムに k 個の初期セントロイドを選ぶ。  
各データポイントを、最も近いセントロイドに割り当てる。
各クラスタ内のデータの平均を計算し、セントロイドを更新。  
セントロイドの位置が変わらなくなるまで、または指定した繰り返し回数まで、ステップ3と4を繰り返す。
## 特徴
計算量: データ数が多いと計算量が増えるが、比較的高速です。  
問題点: クラスタ数kを事前に指定しなければならない。適切なkを選ぶことが難しい場合がある。  
非線形の分布に弱い（円や楕円のような形状のクラスタには強いが、複雑な形状のクラスタには不向き）。

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

# サンプルデータ
X = np.random.rand(100, 2)

# k-meansクラスタリング
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)

# 結果の表示
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='X', color='red')
plt.show()
```
